```python
import urllib3
import os
from dotenv import load_dotenv
import json
from urllib import parse
```

```python
os.chdir("../")
```

```python
load_dotenv('.env.dev')
```

```python
http = urllib3.PoolManager()
```

```python
api_key = os.environ.get('NOOKAL_API_KEY')
```

```python
def get_patients():
    resp = http.request("GET", 
                        f"https://api.nookal.com/production/v2/getPatients?api_key={api_key}")
    data = resp.data.decode('utf-8')  # Decode byte string to a regular string
    json_data = json.loads(data)  # Convert JSON string to a Python dictionary
    return json_data
```

```python
get_patients()
```

```python
def get_appointments():
    resp = http.request("GET", 
                        f"https://api.nookal.com/production/v2/getAppointments?api_key={api_key}")
    data = resp.data.decode('utf-8')  # Decode byte string to a regular string
    json_data = json.loads(data)  # Convert JSON string to a Python dictionary
    return json_data
```

```python
get_appointments()
```

#### Book Appointments

```python
def add_appointment_booking(query_data):
    resp = http.request("GET", 
                        f"https://api.nookal.com/production/v2/addAppointmentBooking?{query_data}",  )
    data = resp.data.decode('utf-8')  # Decode byte string to a regular string
    json_data = json.loads(data)  # Convert JSON string to a Python dictionary
    return json_data
```

```python
query_params = {
    "api_key": api_key,
    "location_id": 1,
    "appointment_date": "2024-03-11",  # Use the correct date
    "start_time": "18:00:00",  # Use the correct start time
    "patient_id": 3,  # Use the actual patient ID
    "practitioner_id": 1,  # Use the actual practitioner ID
    "appointment_type_id": 4,  # Use the actual appointment type ID
    # "notes": "this is a test",  # Any notes you want to include
    # "allow_overlap_bookings": "true"  # Or "false" as required
}

add_appointment_booking(parse.urlencode(query_params))
```
